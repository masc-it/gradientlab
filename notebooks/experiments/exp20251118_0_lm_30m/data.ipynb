```python
from gradientlab.data_utils.text_ds_packing import TextDatasetPacker
from gradientlab.tokenizers.byte_tokenizer import byte_tokenizer
from datasets import load_from_disk
```

```python
tok = byte_tokenizer()
packer = TextDatasetPacker(
    tok,
    bos="<|im_start|>",
    eos="<|im_end|>",
    max_seq_len=512
)
```

```python
ds = load_from_disk("/Users/maurosciancalepore/projects/Autotagging/data.tmp/ETL/gibs_ocr/processed/Gibs2025_V1")
ds
```

```python
ds = packer.pack_dataset(ds, "text")
```

```python
ds.save_to_disk("/Users/maurosciancalepore/projects/Autotagging/data.tmp/ETL/gibs_ocr/processed/Gibs2025_V1_Packed")
```
